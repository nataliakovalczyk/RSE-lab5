```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 1000)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')

plt.xlim(0, 4 * np.pi)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) and cos(x) from 0 to 4π')
plt.legend()
plt.grid(True)
plt.show()
```
